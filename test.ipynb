```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("runs/detect/train10/results.csv")

#Loss results
plt.figure(figsize=(12,5))

train_loss = df['train/box_loss'] + df['train/cls_loss'] + df['train/dfl_loss']
val_loss = df['val/box_loss'] + df['val/cls_loss'] + df['val/dfl_loss']

plt.plot(df['epoch'], train_loss, label='Train Loss', color='blue')
plt.plot(df['epoch'], val_loss, '--', label='Validation Loss', color='orange')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

#Accuracy results
plt.figure(figsize=(12,5))

plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5', color='green')
plt.plot(df['epoch'], df['metrics/precision(B)'], label='Precision', color='blue')
plt.plot(df['epoch'], df['metrics/recall(B)'], label='Recall', color='red')

plt.xlabel("Epoch")
plt.ylabel("Metric Value")
plt.title("Model Accuracy Over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("runs/detect/train10/results.csv") 

final_epoch = df.tail(1) #50th epoch dataframe

precision_val = final_epoch['metrics/precision(B)'].values[0]
recall_val = final_epoch['metrics/recall(B)'].values[0]
map50_val = final_epoch['metrics/mAP50(B)'].values[0]
map50_95_val = final_epoch['metrics/mAP50-95(B)'].values[0]

print("Final Validation Metrics:")
print(f"Precision: {precision_val:.4f}")
print(f"Recall: {recall_val:.4f}")
print(f"mAP@0.5: {map50_val:.4f}")
print(f"mAP@0.5:0.95: {map50_95_val:.4f}")
```

```python
import cv2
from ultralytics import YOLO
model = YOLO("runs/detect/train14/weights/best.pt") 
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    results = model(frame)
    annotated_frame = results[0].plot() 
    cv2.imshow("YOLOv8 Pothole Detection", annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
