```python
import os

labels_dir = r"C:\Users\pra4t\Downloads\Pothole Detection.v10-v5-rotation-only.yolov8\train\labels"

empty_files = []
non_empty_files = []

for file in os.listdir(labels_dir):
    if file.endswith(".txt"):
        file_path = os.path.join(labels_dir, file)
        #print(f"",file_path)
        if os.path.getsize(file_path) == 0:
            empty_files.append(file)#no pothole in  image
        else:
            non_empty_files.append(file) #potholes present

print(f"Total label files: {len(os.listdir(labels_dir))}")
print(f"Images with potholes: {len(non_empty_files)}")
print(f"Images without potholes: {len(empty_files)}")

for empty in empty_files:
    print(empty)
```

```python
from ultralytics import YOLO
model = YOLO("runs/detect/train12/weights/best.pt")
model.train(
    data="data.yaml",
    epochs=120,                 
    imgsz=640,                  
    batch=16,                   
    lr0=0.001,                  
    cos_lr=True,                # smooth learning rate decay
    optimizer="AdamW",          
    patience=40,                
    warmup_epochs=3.0,          
    augment=True,               
    mosaic=0.8,                 
    mixup=0.2,                  
    hsv_h=0.015, hsv_s=0.7, hsv_v=0.4,  # HSV augmentations-hue,saturation,value(brightness)
    translate=0.1, scale=0.5, shear=0.1, flipud=0.0, fliplr=0.5,#translate horizontally and vertically by fraction of imgsz
    freeze=10,#freeze first 10 layers by reducing trainable parameters               
    seed=42,
)
```
